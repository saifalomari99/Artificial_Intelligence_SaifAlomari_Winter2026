```python
import openai 

openai.api_key = "YOUR-IP-KEY-HERE"


# ----- define a helper function that sends a prompt to GPT
def call_llm(prompt: str) -> str:
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a Python coding assistant."},
            {"role": "user", "content": prompt},
        ]
    )
    return response.choices[0].message.content
```

```python
critic_prompt = """
You are a Critic agent whose job is to improve prompts that request Python code.

You will be given [Input Prompt]. Analyze the prompt and rewrite it so that it is more precise, structured, and likely to produce robust and executable Python code.

Your revised prompt must improve the original by making the task clearer and reducing ambiguity. The revised prompt should include concrete requirements that help the model generate working code.

When revising the prompt, ensure that it:

1. Clearly states the programming task and expected output.
2. Specifies the programming language (Python) and appropriate libraries or frameworks.
3. Explains how data should be loaded and preprocessed.
4. Defines model inputs, outputs, and training behavior if machine learning is involved.
5. Requests a complete executable script, not partial code.
6. Includes training, evaluation, and visualization steps when appropriate.
7. Includes comments and clear code structure.
8. Handles missing files, missing data, or empty folders safely.
9. Uses reasonable default assumptions when information is missing.
10. Avoids vague instructions and replaces them with concrete implementation guidance.

Your revision should make the prompt more detailed and robust than the input prompt. If the input prompt is already detailed, improve it further by adding clearer structure, stronger assumptions, and better robustness requirements.

Do not change the main goal of the original prompt.

Output only the revised prompt text.

[Input Prompt]
{input_prompt}
"""
```

```python
user_prompt = """
I have multiple larvae trajectories stored as several CSV files in the csv data folder. Every
0.033333 seconds, the positions of the larva’s head, middle, and tail are recorded. The CSV
columns are time sec, tail x, tail y, mid x, mid y, head x, and head y. I want to train a
vanilla RNN on these trajectories so that the RNN moves like these larvae. Please write the
python code.
"""
```

## First Loop

```python
first_input = critic_prompt.format(input_prompt=user_prompt)

first_revision = call_llm(first_input)

print("First revision:\n")
print(first_revision)
```

## Second loop (Critic improves the improved prompt)

```python
second_input = critic_prompt.format(input_prompt=first_revision)

second_revision = call_llm(second_input)

print("\nSecond revision:\n")
print(second_revision)
```

```python
print("Original Prompt:\n", user_prompt)
print("\nFirst Improved Prompt:\n", first_revision)
print("\nSecond Improved Prompt:\n", second_revision)
```
